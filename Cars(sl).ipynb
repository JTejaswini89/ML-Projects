```python
#loading the dataset
data=data('cars')
data
```

```python
#the first 5 rows from the dataset
data.head()
```

```python
#the basic informtion regarding the dataset
data.info()
```

```python
#checking if null values are in the dataset
data.isnull().sum()
```

```python
data.plot(x='speed',y='dist',style='o')
plt.title("Plotting Distance vs speed")
plt.xlabel("speed")
plt.ylabel("distance")
plt.show()
```

```python
#Features and labels or dependent and inpendent columns from dataset
x=data.drop('dist',axis=1)
y=data['dist']
```

```python
#splltting dataset for trainign and testing
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
x_train.shape
```

```python
y_train.shape
```

```python
#build the model
linear=LinearRegression()
linear.fit(x_train,y_train)
```

```python
#Predicting the y values from testing dataset
y_pred=linear.predict(x_test)
y_pred
```

```python
print("coefficient:",linear.coef_)
print("intercet:",linear.intercept_)
```

```python
#plottign line equation
plt.scatter(x,y)
plt.plot(x_test,y_pred,color='red')
plt.show()
```

```python
#model evaluation
print("Mean squared error:",mean_squared_error(y_test,y_pred))
print("R2 score:",r2_score(y_test,y_pred))
```

```python
#custom input prediction
custom_speed=np.array(4)
predicted_dist=linear.predict(custom_speed.reshape(1,-1))
print("the distnce travelled is:",predicted_dist)
```
